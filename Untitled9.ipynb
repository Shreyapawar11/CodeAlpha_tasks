```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
data = pd.read_csv('Unemployment in India.csv')
```

```python
print(data.head())
```

```python
data['Date'] = pd.to_datetime(data[' Date'])
data.set_index('Date', inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data.fillna(method='ffill', inplace=True)
```

```python
print(data.columns)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[' Estimated Unemployment Rate (%)'], label='Estimated Unemployment Rate (%)', color='blue')
plt.title('Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data[' Estimated Unemployment Rate (%)'], bins=30, kde=True)
plt.title('Distribution of Unemployment Rates')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
plot_acf(data[' Estimated Unemployment Rate (%)'])
plt.title('Autocorrelation Function')
plt.show()
```

```python
plot_pacf(data[' Estimated Unemployment Rate (%)'])
plt.title('Partial Autocorrelation Function')
plt.show()
```

```python
model = ARIMA(data[' Estimated Unemployment Rate (%)'], order=(1, 1, 1))
model_fit = model.fit()
```

```python
print(model_fit.summary())
```

```python
print(data.index[-1])
print(pd.isnull(data.index[-1]))
```

```python
forecast = model_fit.forecast(steps=12)


if pd.isnull(data.index[-1]):
    last_valid_date = data.index[data.index.notnull()][-1]
    forecast_index = pd.date_range(start=last_valid_date + pd.DateOffset(months=1), periods=12, freq='MS')
else:
    forecast_index = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[' Estimated Unemployment Rate (%)'], label='Historical Unemployment Rate', color='blue')
plt.plot(forecast_index, forecast, label='Forecasted Unemployment Rate', color='orange')
plt.title('Unemployment Rate Forecast')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid()
plt.show()
```
